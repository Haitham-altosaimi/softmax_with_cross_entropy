```python
import numpy as np
import numpy.random as random
```

```python
def Normailze_Data(data):
  return (data-np.min(data))/(np.max(data)-np.min(data))
```

```python
x = np.vstack( (random.randint(0,10, (10,2)), 
               random.randint(10,20, (10,2)),
               (random.randint(20,30,(10,2)))) )
x = Normailze_Data(x)

y = np.vstack( ( np.array(([1,0,0])*10).reshape(10,3),
                np.array(([0,1,0])*10).reshape(10,3),
                np.array(([0,0,1])*10).reshape(10,3)))
```

```python
def forward_pass(x, w1, b1, w2, b2, y):
  hidden_1 = x.dot(w1.T) + b1
  hidden_1_sig = 1 / (1 + np.exp(-hidden_1))
  hidden_2 = hidden_1_sig.dot(w2) + b2
  hidden_2_soft = np.exp(hidden_2)/np.exp(hidden_2).sum(axis=1, keepdims=True)
  Loss = -np.mean(np.multiply(y, np.log(hidden_2_soft)))
  print(" the loss is: ", Loss.sum())
  return hidden_1, hidden_1_sig, hidden_2, hidden_2_soft


```

```python
def backward_pass(x, hidden_2_soft, hidden_2, hidden_1_sig, hidden_1, y):
  dLdw2 = ((hidden_2_soft-y).T.dot(hidden_1_sig)).T
  dLdw1 = (((hidden_2_soft-y).dot(w2.T))*(np.multiply(hidden_1_sig, (1-hidden_1_sig)))).T.dot(x)

  dLdb2 = (hidden_2_soft - y).sum(axis=0)
  dLdb1 = ((hidden_2_soft - y).dot(w2.T)*(np.multiply(hidden_1_sig, (1-hidden_1_sig)))).sum(axis=0)
  return dLdw2, dLdw1, dLdb2, dLdb1



```

```python
def update_parameters(w1, w2, dw1, dw2, b1, b2, db1, db2, learning_rate):
  new_w1 = w1 - dw1 * learning_rate
  new_w2 = w2 - dw2 * learning_rate
  new_b1 = b1 - db1 * learning_rate
  new_b2 = b2 - db2 * learning_rate
  return new_w1, new_w2, new_b1, new_b2
```

```python
w1 = random.uniform(-1,1,(100,2))
b1 = random.uniform(-1,1,(1,100))
w2 = random.uniform(-1,1,(100,3))
b2 = random.uniform(-1,1,(3))
```

```python
for i in range(100):
  hidden_1, hidden_1_sig, hidden_2, hidden_2_soft = forward_pass(x, w1, b1, w2, b2, y)
  dw2, dw1, db2, db1= backward_pass(x, hidden_2_soft, hidden_2, hidden_1_sig, hidden_1, y)
  learning_rate = 1e-3
  new_w1, new_w2, new_b1, new_b2 = update_parameters(w1, w2, dw1, dw2, b1, b2, db1, db2, learning_rate)
  w1, w2, b1, b2 = new_w1, new_w2, new_b1, new_b2






```
